```python
class Robot:
    def __init__(self, name):
        self.name = name
        print("Robot init'd")        
        
    def get_info(self):
        print("Hi, I am " + self.name)
        
class TallRobot(Robot):
    def __init__(self, name, size, **kwargs):
        print("TallRobot init")
        print(kwargs)
        self.size = size        
        super().__init__(name, **kwargs)
#         Robot.__init__(self, name)
        print("TallRobot init'd")        
        
    def get_info(self):
        print("Hi, I am " + self.name + " and I am " + str(self.size) + " meter tall.")
        
class HeavyRobot(Robot):
    def __init__(self, name, weight, **kwargs):
        print("HeavyRobot init")
        print(kwargs)        
        self.weight = weight        
        super().__init__(name, **kwargs)
#         Robot.__init__(self, name)
        print("HeavyRobot init'd")     
        
    def get_info(self):
        print("Hi, I am " + self.name + " and I weigh " + str(self.weight) + " kg.")  
        
class DoomsdayRobot(TallRobot, HeavyRobot):
    def __init__(self, name, size, weight):
        print("DoomsdayRobot init")
        super().__init__(name, size=size, weight=weight)
#         TallRobot.__init__(self, name, size)
#         HeavyRobot.__init__(self, name, weight)
        print("DoomsdayRobot init'd")     
        
    def get_info(self):
        print("Name: " + self.name + "; weight: " + str(self.weight) + "; size: " + str(self.size))          
        
class ColoredDoomsdayRobot(DoomsdayRobot):
    def __init__(self, name, size, weight, color):
        print("ColoredDoomsdayRobot init")
        self.color = color
        super().__init__(name, size=size, weight=weight)
#         DoomsdayRobot.__init__(self, name, size, weight)
        print("ColoredDoomsdayRobot init'd")
    
    def get_info(self):
        print("Name: " + self.name + "; weight: " + str(self.weight) +
              "; size: " + str(self.size) + "; color: " + self.color)     
        
r = Robot('James')
print(r.name)
r.get_info()

tall_r = TallRobot('Peter',3)
tall_r.get_info()

heavy_r = HeavyRobot('Matthew',200)
heavy_r.get_info()

dday_r = DoomsdayRobot('John',2.5, 180)
dday_r.get_info()

c_r = ColoredDoomsdayRobot('Paul',3.5, 220,'red')
c_r.get_info()
```

`super()` is closely connected to the concepts of the [method resolution order](https://docs.python.org/3/glossary.html#term-method-resolution-order) (see the `__mro__` attribute).

`super()` is great to call a function defined by a parent class. But it is most useful in cases of multiple inheritance.

```python
class a:
    z = 1
    def __init__(self):
        self.x = 2
                
    def i_am_a():
        print("class a")
        
    def who_am_i(self):
        print("class a")
        
class b(a):
    def __init__(self):
        self.y = 3
        
    # "overriding" who_am_i
    def who_am_i(self):
        print("class b")
        
i = a()
print(i.x)
print(i.z)
a.i_am_a()
i.who_am_i()

j = b()
print(j.y)
print(j.z)
j.who_am_i()
```
